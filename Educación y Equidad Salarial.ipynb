```python
# 1. Importar librer√≠as esenciales
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Definir la ruta del archivo
ruta = r"C:\Users\josue\OneDrive\Escritorio\mi_proyecto\dataset_empleados.csv"
# 3. Cargar el dataset en un DataFrame
df = pd.read_csv(ruta)

# 4. Revisar las primeras filas para confirmar la carga
df.head()
```

```python
list(df.columns)
```

```python
# Calcular la media de la edad
media_edad = df["Edad"].mean()
print("La media de la edad es:", media_edad)
```

```python
import matplotlib.pyplot as plt

# Histograma de la columna Edad
plt.figure(figsize=(8,5))
plt.hist(df["Edad"], bins=10, color="skyblue", edgecolor="black")

# Etiquetas y t√≠tulo
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de la Edad de Empleados")

plt.show()
```

```python
# Calcular la mediana de la edad
mediana_edad = df["Edad"].median()
print("La mediana de la edad es:", mediana_edad)
```

```python
#üîç Diferencia clave entre media y mediana
Media (promedio): suma de todas las edades dividido entre el n√∫mero de empleados.

Mediana: valor central cuando las edades est√°n ordenadas.

Si el n√∫mero de empleados es impar ‚Üí la edad del centro.

Si es par ‚Üí el promedio de las dos edades centrales.
```

```python
# Calcular la moda de la edad
moda_edad = df["Edad"].mode()
print("La moda de la edad es:", moda_edad[0])
```

```python
# La moda representa el valor m√°s frecuente en la columna Edad
```

```python
# Promedio de salario seg√∫n nivel educativo
df.groupby("NivelEducativo")["Salario"].mean()
```

```python
#Esto te mostrar√° cu√°nto gana en promedio cada grupo educativo.
```

```python
# Boxplot para ver la distribuci√≥n de salarios por nivel educativo
plt.figure(figsize=(8,5))
sns.boxplot(x="NivelEducativo", y="Salario", data=df, hue="NivelEducativo", palette="Set2", legend=False)
plt.title("Distribuci√≥n de Salarios seg√∫n Nivel Educativo")
plt.show()
```

```python
#La l√≠nea del centro de cada caja representa el salario m√°s com√∫n (la mediana).

#La altura de la caja muestra d√≥nde se concentra la mayor√≠a de los salarios.

#Las l√≠neas que salen de la caja indican los valores m√°s bajos y m√°s altos dentro de lo normal.

#Si ves puntos fuera de esas l√≠neas, son salarios que se salen del promedio: pueden ser muy altos o muy bajos.

```
